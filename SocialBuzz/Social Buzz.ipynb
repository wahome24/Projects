#### The purpose of the below code is to analyse the top five content categories based on reaction score.

```python
import numpy as np
import pandas as pd
```

```python
#Importing dataset

content = pd.read_csv(r"C:\Users\USER\Desktop\Virtual Internships\Data.csv")
```

```python
#Displaying the data

content
```

```python
#Updating the Type column name

content.columns
```

```python
content.columns = ['Content ID', 'Reaction Type', 'Reaction Score', 'Category']
```

```python
content
```

```python
#New dataframe to store relevant columns for our analysis

df = content[['Category','Reaction Type','Reaction Score']]
```

```python
df
```

```python
#Checking column data types

df.dtypes
```

```python
df['Category1']= df['Category'].str.lower()
```

```python
df
```

```python
#Calculating total review scores for each category

df.groupby('Category1')['Reaction Score'].sum()
```

```python
#Storing the grouped categories as a data frame in order to perform sorting

result = pd.DataFrame(df.groupby('Category')['Reaction Score'].sum())
```

```python
result
```

```python
top_five = result.sort_values('Reaction Score', ascending = False )
```
